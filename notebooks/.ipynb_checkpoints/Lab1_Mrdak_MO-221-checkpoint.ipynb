## Учёт инфляции

Перепишим код из .pdf файла. Проблемы могут возникнуть с фильмами, дата производства которых неизвестна, поэтому реальный бюджет, сборы и т.д. занулим.

```python
from bs4 import BeautifulSoup
import requests

url = "https://www.usinflationcalculator.com/inflation/consumer-price-index-and-annual-percent-changes-from-1913-to-2008/"
r = requests.get(url)
data = r.text
soup = BeautifulSoup(data, 'html.parser')

table = soup.find("table")
rows = table.tbody.findAll("tr")

years = []
cpis = []

for row in rows:
    year = row.findAll('td')[0].get_text()
    if year.isdigit() and int(year) < 2017:
        years.append(int(year))
        cpis.append(float(row.findAll('td')[13].get_text()))

cpi_table = pd.DataFrame({
    'year': years,
    'avg_annual_cpi': cpis
})

cpi_table
```

```python
def get_real_value(nominal_amt, old_cpi, new_cpi):
    real_value = (nominal_amt * new_cpi) / old_cpi
    return real_value
```

```python
CPI_2016 = float (cpi_table[cpi_table['year'] == 2016]['avg_annual_cpi'].iloc[0])
```

```python
real_domestic_gross = []
real_budget_values = []

for index, row in df.iterrows():
    gross = row['gross']
    budget = row['budget']
    year = row['title_year']
    if(year!=0):
        cpi = float(cpi_table[cpi_table['year'] == int(year)]['avg_annual_cpi'].iloc[0])
        real_gross = get_real_value(gross, cpi, CPI_2016)
        real_budget = get_real_value(budget, cpi, CPI_2016)
    else:
        real_gross = 0
        real_budget = 0
    real_domestic_gross.append(real_gross)
    real_budget_values.append(real_budget)

df['real_domestic_gross'] = real_domestic_gross
df['real_budget_values'] = real_budget_values
```

Здесь может быть проблема с фильмами, у которых нулевой бюджет, потому что мы делим на него при подсчёте roi, поэтому для таких фильмов я roi приравниваю к 0.

```python
profits = []
roi_vals = []

for index, row in df.iterrows():
    profit = row['real_domestic_gross'] - row['real_budget_values']
    budget = row['real_budget_values']
    if(budget!=0):
        num = profit - budget
        den = budget
        roi = (num/den)*100
    else:
        roi = 0
    profits.append(profit)
    roi_vals.append(roi)
    
df['profit'] = profits
df['roi'] = roi_vals
```

### Гипотеза №1.

 #### Чем больше бюджет у фильмов Нолана, тем больше их оценка на imdb

Создадим сводную таблицу, где индексами укажем имя режиссёра и название фильма, а значениями - реальный бюджет и оценку на imdb. Затем отсортируем таблицу по уменьшению бюджета и отберём только те фильмы, режиссёром которых является Кристофер Нолан 

```python
pvt = df.pivot_table(index=['director_name','movie_title'], values=['real_budget_values', 'imdb_score'])
pvt = pvt.sort_values(by=['real_budget_values'], ascending=False)
pvt = pvt.query('director_name == ["Christopher Nolan"]')
pvt.head(20)
```

Гипотеза не подтвердилась, уже у второго фильма бюджет меньше, но оценка выше

### Гипотеза №2.

 #### Число зрительских оценок австралийских фильмов зависит от числа лайков фильма в Facebook, начиная с 2010 года

Создадим сводную таблицу, где индексами укажем страну производства, название фильма и год выпуска, а значениями - число лайков фильма в Facebook и число зрителей, которые поставили оценку. Затем отсортируем таблицу по уменьшению числа лайков и отберём только те фильмы, которые сняты в Австралии и с 2010 года

```python
pvt_2 = df.pivot_table(index=['country','movie_title','title_year'], values=['movie_facebook_likes', 'num_voted_users'])
pvt_2 = pvt_2.sort_values(by=['movie_facebook_likes'], ascending=False)
pvt_2 = pvt_2.query('country == ["Australia"]')
pvt_2 = pvt_2.query('title_year >= 2010')
pvt_2.head(10)
```

Гипотеза не подтвердилась. Если сравнивать показатели четвртого и пятого фильмов, становится очевидно, что гипотеза неверна. 

### Гипотеза №3

#### В США больше половины фильмов сняты в жанре "драма", а в Японии в жанре "приключение"

Создадим два датафрейма. В первом будет информация только о тех фильмах, что были сняты в США, а во втором только те, что были сняты в Японии

```python
df_japan = df.loc[df.loc[:,'country'] == 'Japan']
df_usa = df.loc[df.loc[:,'country'] == 'USA']
```

Сгруппируем фильмы. В первой группе будут только те, что имеют среди жанров драму, а во второй только те, что не имеют данного жанра. Затем посчитаем число фильмов, входящих в каждую из групп, и отсортируем по убыванию.

```python
usa = df_usa.groupby(df['genres'].str.contains('Drama'))['movie_title'].count().sort_values(ascending=False)
usa
```

Для фильмов из Японии сделаем то же самое.

```python
japan = df_japan.groupby(df['genres'].str.contains('Adventure'))['movie_title'].count().sort_values(ascending=False)
japan
```

Гипотеза оказалась верна лишь частично. В Японии из 22 фильмов 13 сняты в жанре приключение, а это больше половины. В США же из 3773 фильмов лишь 1837 имеют в жанрах драму.
